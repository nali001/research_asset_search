# Export ðŸ¤– images w/ pyvips

```python
ls_imgs = glob.glob(os.path.join(DATASET_FOLDER, "train", "*.tif"))
```

```python
import gc
import pyvips
from tqdm.auto import tqdm

for tif_path in tqdm(ls_imgs):
    # img = Image.open(tif_path)
    name, _ = os.path.splitext(os.path.basename(tif_path))
    png_path = os.path.join("./train_images", f"{name}.png")
    pyvips.Image.thumbnail(tif_path, 2000).write_to_file(png_path)
    gc.collect()
```

# Show ðŸ”Ž few images

```python
df_train = pd.read_csv(os.path.join(DATASET_FOLDER, "train.csv"))
labels = list(df_train["label"].unique())
display(df_train.head())
```

```python
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(16, 20))
for lb, dfg in df_train.groupby("label"):
    j = labels.index(lb)
    for i, name in enumerate(dfg["image_id"].sample(5)):
        # print(f"{lb}: {name}")
        img_path = os.path.join("train_images", f"{name}.png")
        axes[i, j].imshow(plt.imread(img_path))
        axes[i, j].set_title(f"{lb}: {name}")
```
