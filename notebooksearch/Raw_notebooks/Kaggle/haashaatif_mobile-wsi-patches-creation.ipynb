```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import cv2
import zipfile
from matplotlib import pyplot as plt
from tqdm.notebook import tqdm
base_path = '../input/mobilewsis/mobilewsis/'
all_cases = os.listdir(base_path)
```

```python
def plot_images(imgs):
    
    for i in range(len(imgs)): 
        plt.figure(figsize=(5, 5))
        plt.figure(i+1)
        if len(imgs[i].shape)<3:
            plt.imshow(imgs[i], cmap="gray")
        else:
            plt.imshow(imgs[i])
```

```python
original_image = cv2.imread("../input/mobilewsis/mobilewsis/Case 1/IMG_4723.jpg")
print("Original Image Shape:",original_image.shape)
resized_image = cv2.resize(original_image, (4000,4000), interpolation = cv2.INTER_AREA)
print("Resized Image Shape:",resized_image.shape)
```

```python
plt.imshow(original_image)
```

```python
plt.imshow(resized_image)
```

```python
height,width,channels = resized_image.shape
print(height,width,channels)
plt.imshow(resized_image[0:1000,0:1000,:])
```

```python
patch_size = 2000
patches = []
for y in range(0,height,patch_size):
    for x in range(0,width,patch_size):
        patches.append(resized_image[y:y+patch_size,x:x+patch_size,:])
print("Total Patches Created:",len(patches))
```

```python
plot_images([resized_image])
```

```python
plot_images(patches)
```
