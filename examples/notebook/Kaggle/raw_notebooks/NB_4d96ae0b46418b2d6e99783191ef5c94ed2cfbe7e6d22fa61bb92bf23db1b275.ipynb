**Machine learning algorithm**

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn import tree
```

**labeling the data for machine learning algorithm**
I will rely on the column "qualsleep4gp" as it does directly label participants as having or not having sleep problems by 1 to 4 .

we take every column that does not address a reason for why the people who are not geeting good sleep are getting it.

```python
#removing labels
Y=sleep_data["qualsleep4gp"].copy()
X=sleep_data.drop(["lethargy","troublestaysleep","energy","sleepy","tired","ess","troublefallasleep","sleepy","staysleeprec","getsleeprec","problem","qualsleep","qualsleep4gp","probsleeprec","totsas","refreshed","driveaccident","satissleep","drivesleep","drivesleeprec"],axis=1)
```

before going into the machine learning to decide which factors affects the sleep problems , I would like to normalize data as some columns like height and age are having different scales so that no certain values dominate the learning process

```python
#data normalization#
# copy the data
df_min_max_scaled = X.copy()

# apply normalization techniques
for column in df_min_max_scaled.columns:
	df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())	

# view normalized data
print(df_min_max_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split( 
          X, Y, test_size = 0.2, random_state = 100)
```

```python
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,max_depth=5, min_samples_leaf=5)
```

```python
History=clf_gini.fit(X_train, y_train)
```

```python
y_pred = History.predict(X_test)
print("Predicted values:")
print(y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.gcf().set_size_inches(14,7)
plt.xticks(rotation = 90) # Rotates X-Axis Ticks by 45-degrees
plt.show()
```
