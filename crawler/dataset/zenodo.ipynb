# Search datasets on Zenodo
Ref: https://help.zenodo.org/guides/search/

```python
import requests

def search_zenodo_datasets(query, size):
    base_url = "https://zenodo.org/api/records/"
    params = {
        "q": query,
        "size": size,  # Number of results to retrieve, adjust as needed
        "type": "dataset", 
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Check for HTTP errors

        # Parse the JSON response
        data = response.json()
        hits = data.get("hits", {}).get("hits", [])
    except requests.exceptions.RequestException as e:
        print("Error:", e)

    return hits

if __name__ == "__main__":
    query = "citeseer"  # Replace with your search query
    hits = search_zenodo_datasets(query, 10)
```
